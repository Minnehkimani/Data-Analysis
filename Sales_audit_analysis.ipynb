```python

###Q1.Check the data to answer the following questions: Which top 5 products contribute 80% of revenue?
#the total in sales items table is equal to revenue
sale_items["revenue"] = sale_items["total"]
```

```python
products = pd.read_csv("/Products _2024.csv")
```

```python
merged = sale_items.merge(products,
                          left_on="product_code",
                          right_on="code",
                        how="left")
#to matche records using product_code from the sales table and code from the products table
```

```python
product_revenue = merged.groupby("label")["revenue"].sum().sort_values(ascending=False)
product_revenue.head(10)
```

```python
#Apply 80/20 Rule (Pareto Analysis)

total_revenue = product_revenue.sum()

cumulative_revenue = product_revenue.cumsum()
cumulative_percentage = (cumulative_revenue / total_revenue) * 100

pareto_df = pd.DataFrame({
    "Revenue": product_revenue,
    "Cumulative %": cumulative_percentage
})

pareto_df.head(10)
```

```python
#to get products that Make Up 80%

top_80 = pareto_df[pareto_df["Cumulative %"] <= 80]
top_80.head(10)
```

```python
###At what time do we experience the highest peak in sales?

#to convert it to pandas datetime


sale_items["created_at"] = pd.to_datetime(sale_items["created_at"])
```

```python
sale_items["hour"] = sale_items["created_at"].dt.hour
```

```python


hourly_sales = sale_items.groupby("hour")["revenue"].sum().sort_values(ascending=False)

hourly_sales.head()
```

```python
### Which products contribute the highest profit margins?
#Profit = Revenue - Cost
#margin percent= profit/revenue
merged["profit"] = merged["revenue"] - merged["unit_cost"]
```

```python
# to aggregate revenue and profit per product, calculate profit margin percentage
product_profit = merged.groupby("label").agg({
    "revenue": "sum",
    "profit": "sum"
})

product_profit["profit_margin_%"] = (product_profit["profit"] / product_profit["revenue"]) * 100

product_profit.sort_values("profit_margin_%", ascending=False).head()
```

```python
###The peak months in terms of sales, and which products were sold the most.
sale_items["month"] = sale_items["created_at"].dt.month
```

```python
#to get monthly revenue
monthly_sales = sale_items.groupby("month")["revenue"].sum().sort_values(ascending=False)

monthly_sales
```

```python
top_products_quantity = merged.groupby("label")["quantity"].sum().sort_values(ascending=False)

top_products_quantity.head(10)
```

```python
# Final cleaned datasets

products = pd.read_csv("/Products _2024.csv")
sale_items = pd.read_csv("/Sale _Items.csv")

products["label"] = products["label"].str.lower().str.strip()
sale_items["price"] = pd.to_numeric(sale_items["price"], errors="coerce")

merged = sale_items.merge(
    products,
    left_on="product_code",
    right_on="code",
    how="left"
)

merged["revenue"] = merged["total"]
merged["profit"] = merged["revenue"] - merged["unit_cost"]

merged.to_csv("cleaned_data.csv", index=False)
```
